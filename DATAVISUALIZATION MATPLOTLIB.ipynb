```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
```

```python
x = np.arange(1,11)
x
```

```python
y1=2*x
y2=3*x

y1,y2
```

```python
plt.plot(x,y1,color='Yellow',linewidth=2,linestyle=':')
plt.plot(x,y2,color='Black',linewidth=2,linestyle=':')
plt.title("Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.show()
```

```python
# BAR CHART
```

```python
student = {"Sam":30,"Bob":50,"Julia":70}
```

```python
names = list(student.keys())
marks = list(student.values())

# Print
names,marks
```

```python
plt.barh(names,marks, color='grey')
plt.title("Bar Plot")
plt.xlabel("Names of Students")
plt.ylabel("Marks")
plt.grid(True)
plt.show()

# barh le horizontal banako bar le vertical hunxa
```

```python
x=[4,7,3,9,1,6]
y1=[9,1,2,6,4,9]
y2=[4,7,8,1,2,6]
plt.scatter(x,y1)
plt.scatter(x,y2,color='red')
plt.grid(True)
```

```python
# PIECHART
```

```python
fruits =["apple","banana", "orange", "mango"]
```

```python
cost =[76,77,45,90]
```

```python
plt.figure(figsize=(10,10))
plt.pie(cost,labels=fruits, autopct='%0.2f%%' , shadow=True)
plt.show()

# 2 ta value dekhako 0.2% wala ma point paxi
```
